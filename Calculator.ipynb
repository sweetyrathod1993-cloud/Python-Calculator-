```python
import tkinter as tk

win = tk.Tk()
win.title("Calculator")
win.configure(bg="#2b2b2b")

# ===== Frames =====
calc_frame = tk.Frame(win, bg="#2b2b2b")
loan_frame = tk.Frame(win, bg="#2b2b2b")

calc_frame.grid(row=0, column=0, sticky="nsew")
loan_frame.grid(row=0, column=0, sticky="nsew")

# ===== Switch Functions =====
def show_calc():
    calc_frame.tkraise()
    entry.focus_set()

def show_loan():
    loan_frame.tkraise()

# ===== Display =====
entry = tk.Entry(
    calc_frame,
    font=("Arial", 24),
    bd=0,
    bg="#1e1e1e",
    fg="white",
    justify="right"
)
entry.grid(row=0, column=0, columnspan=5, padx=10, pady=(10, 10), ipady=15, sticky="nsew")

# ===== History =====
history_box = tk.Text(
    calc_frame,
    height=6,
    bg="#1e1e1e",
    fg="#9cdcfe",
    font=("Arial", 10),
    bd=0
)
history_box.grid(row=1, column=0, columnspan=5, padx=10, pady=(0, 10), sticky="nsew")
history_box.config(state="disabled")

# ===== Calculator Functions =====
def click(val):
    entry.insert(tk.END, val)

def clear():
    entry.delete(0, tk.END)

def backspace():
    if len(entry.get()) > 0:
        entry.delete(len(entry.get())-1, tk.END)

def add_history(expression, result):
    history_box.config(state="normal")
    history_box.insert(tk.END, f"{expression} = {result}\n")
    history_box.see(tk.END)
    history_box.config(state="disabled")

def equal():
    try:
        expr = entry.get()
        expr = expr.replace('x', '*').replace('mod', '%').replace('pow', '**')
        result = str(eval(expr))
        add_history(expr, result)
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

# ===== Keyboard Support =====
def key_event(event):
    key = event.keysym
    if key in ["Return", "KP_Enter"]:  # ✅ ENTER FIX
        equal()
        return "break"

# Bind globally
win.bind("<KeyPress>", key_event)

# ===== Buttons =====
def create_button(text, row, col, cmd, bg="#3c3f41"):
    tk.Button(calc_frame, text=text, bg=bg, fg="white",
              font=("Arial", 14, "bold"), bd=0,
              command=cmd).grid(row=row, column=col, padx=5, pady=5, sticky="nsew")

create_button("C", 2, 0, clear, "#d9534f")
create_button("⌫", 2, 1, backspace, "#f0ad4e")
create_button("(", 2, 2, lambda: click("("))
create_button(")", 2, 3, lambda: click(")"))
create_button("/", 2, 4, lambda: click("/"), "#f0ad4e")

create_button("7", 3, 0, lambda: click("7"))
create_button("8", 3, 1, lambda: click("8"))
create_button("9", 3, 2, lambda: click("9"))
create_button("*", 3, 3, lambda: click("*"), "#f0ad4e")

create_button("4", 4, 0, lambda: click("4"))
create_button("5", 4, 1, lambda: click("5"))
create_button("6", 4, 2, lambda: click("6"))
create_button("-", 4, 3, lambda: click("-"), "#f0ad4e")

create_button("1", 5, 0, lambda: click("1"))
create_button("2", 5, 1, lambda: click("2"))
create_button("3", 5, 2, lambda: click("3"))
create_button("+", 5, 3, lambda: click("+"), "#f0ad4e")

tk.Button(calc_frame, text="=", bg="#5cb85c", fg="white",
          font=("Arial", 14, "bold"), bd=0,
          command=equal).grid(row=5, column=4, rowspan=2, padx=5, pady=5, sticky="nsew")

create_button("0", 6, 0, lambda: click("0"))
create_button(".", 6, 1, lambda: click("."))
create_button("%", 6, 2, lambda: click("%"))
create_button("pow", 6, 3, lambda: click("**"))

# ===== Loan UI =====
tk.Label(loan_frame, text="Loan EMI Calculator",
         bg="#2b2b2b", fg="white", font=("Arial", 16)).grid(row=0, column=0, columnspan=2, pady=10)

# Loan Amount
tk.Label(loan_frame, text="Loan Amount",
         bg="#2b2b2b", fg="white", font=("Arial", 11)).grid(row=1, column=0, padx=10, pady=5, sticky="w")

loan_amount = tk.Entry(loan_frame)
loan_amount.grid(row=1, column=1, padx=10, pady=5)

# Interest Rate
tk.Label(loan_frame, text="Interest Rate",
         bg="#2b2b2b", fg="white", font=("Arial", 11)).grid(row=2, column=0, padx=10, pady=5, sticky="w")

interest_rate = tk.Entry(loan_frame)
interest_rate.grid(row=2, column=1, padx=10, pady=5)

# Tenure Month
tk.Label(loan_frame, text="Tenure Month",
         bg="#2b2b2b", fg="white", font=("Arial", 11)).grid(row=3, column=0, padx=10, pady=5, sticky="w")

tenure = tk.Entry(loan_frame)
tenure.grid(row=3, column=1, padx=10, pady=5)

# Result Label
result_label = tk.Label(loan_frame, text="", bg="#2b2b2b", fg="#9cdcfe")
result_label.grid(row=4, column=0, columnspan=2, pady=10)

def calculate_emi():
    try:
        P = float(loan_amount.get())
        R = float(interest_rate.get()) / 12 / 100
        N = int(tenure.get())
        emi = (P * R * (1 + R)**N) / ((1 + R)**N - 1)
        result_label.config(text=f"EMI: {round(emi, 2)}")
    except:
        result_label.config(text="Error")

tk.Button(loan_frame, text="Calculate EMI",
          command=calculate_emi,
          bg="#5cb85c", fg="white").grid(row=4, column=0, columnspan=2, pady=0)

tk.Button(loan_frame, text="⬅ Back",
          command=show_calc).grid(row=4, column=2, padx=2, pady=2)

# ===== Menu =====
menu_bar = tk.Menu(win)
win.config(menu=menu_bar)

loan_menu = tk.Menu(menu_bar, tearoff=0)
menu_bar.add_cascade(label="Loan", menu=loan_menu)

loan_menu.add_command(label="EMI Calculator", command=show_loan)

# ===== Layout =====
for i in range(7):
    calc_frame.grid_rowconfigure(i, weight=1)

for i in range(5):
    calc_frame.grid_columnconfigure(i, weight=1)

# Default screen
show_calc()

entry.focus_set()
win.mainloop()
```
